```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
import pandas as pd
df= pd.read_csv("/content/creditcard.csv")
df.head()
```

```python
df.tail()
```

```python
# Dataset information
df.info()
```

```python
# checking the number of missing values in each column
df.isnull().sum()
```

```python
# distribution of legit transaction and fraudulent transaction
df['Class'].value_counts()
```

This Dataset is highly unblanced

0 ---> Normal Transaction

1 ---> Fraudulent Transaction



```python
# separating the data for analysis
legit = df[df.Class == 0]
fraud = df[df.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
# statistical measures of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# compare the values for both transaction
df.groupby('Class').mean()
```

### Under-Sampling


-->> Build a sample dataset containing similar distribution of normal transactions an Fraudulent Transaction


-->> Number of Fraudulent Transaction --> 393

```python
legit_sample = legit.sample(n=393)
```

**Concatenating Two DataFrames**

```python
new_dataset = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

Splitting the data into Features & targets

```python
X = new_dataset.drop(columns='Class', axis=1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

**Split the data into Training data and testing data **

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

**Model Training**

Logistic Regression

```python
model = LogisticRegression()
```

```python
# taining the Logistic Regression Model with training data
model.fit(X_train, Y_train)
```

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training data : ', training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy on Test data : ', test_data_accuracy)
```
